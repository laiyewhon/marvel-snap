```python
import requests
import hashlib
import json
import pandas as pd

from functools import reduce

r = requests.get('https://marvelsnapzone.com/getinfo/?searchtype=cards&searchcardstype=true')
```

Add `cid` to `tags`

```python
cleansed_json = r.json()['success']['cards']

for jcard in cleansed_json:
    cid = jcard['cid']
    for jtags in jcard['tags']:
        jtags.update({ 'cid': cid })
```
